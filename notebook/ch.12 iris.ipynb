```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import tensorflow as tf
import seaborn as sns
from sklearn.preprocessing import LabelEncoder

%matplotlib inline
```

```python
df = pd.read_csv("https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv",
                names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species'], skiprows=1)
```

```python
df.head(10)
```

```python
sns.pairplot(df, hue='species')
plt.show()
```

Y 값이 문자열

* Setosa
* Versicolor
* Virginica

로 되어 있으므로 `One-hot-encoding` 을 이용해

* `[1, 0, 0]`
* `[0, 1, 0]`
* `[0, 0, 1]`

형태로 바꿔주는 작업이 필요함

```python
dataset = df.values
X = dataset[:, 0:4].astype(float)
Y_obj = dataset[:, 4]
```

```python
Y_obj
```

```python
e = LabelEncoder()
e.fit(Y_obj)
Y = e.transform(Y_obj)
Y
```

```python
Y_encoded = tf.keras.utils.to_categorical(Y)
Y_encoded
```

```python
model = Sequential()
model.add(Dense(16, input_dim=4, activation='relu'))
model.add(Dense(3, activation='softmax')) # 활성함수로 softmax 사용
```

## Softmax

입력값에서 계산된 각각의 가중 합의 결과를
1이 되어지도록 변환하는 과정

큰값과 작은값이 서로 더 크고 작게 두드러짐

이를 이용해 원핫 레이블에 도달하면

* Setosa
* Versicolor
* Virginica

## 모델링

```python
np.random.seed(3)
tf.random.set_seed(3)
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X, Y_encoded, epochs=50, batch_size=1)
```

```python
model.evaluate(X, Y_encoded)[1]
```
